```python
# Import required libraries
from textblob import TextBlob
import matplotlib.pyplot as plt
```

```python
# Sample data: List of social media posts (e.g., tweets)
data = [
    "I love data science! It's amazing.",
    "Data science is difficult, but rewarding.",
    "I hate working with big data, it's frustrating.",
    "Machine learning is a cool field in data science.",
    "I am tired of all the data wrangling!",
    "Data science is the future of technology."
]
```

```python
# Function to analyze sentiment of text using TextBlob
def get_sentiment(text):
    # Perform sentiment analysis and return polarity
    analysis = TextBlob(text)
    return analysis.sentiment.polarity
```

```python
# List to store sentiment values
sentiment_scores = [get_sentiment(text) for text in data]
```

```python
# Categorize sentiments as Positive, Neutral, or Negative
positive = len([score for score in sentiment_scores if score > 0])
neutral = len([score for score in sentiment_scores if score == 0])
negative = len([score for score in sentiment_scores if score < 0])
```

```python
# Visualization: Pie chart of sentiment distribution
labels = ['Positive', 'Neutral', 'Negative']
sizes = [positive, neutral, negative]
colors = ['lightgreen', 'lightblue', 'lightcoral']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title("Sentiment Analysis of Social Media Posts")
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
plt.show()
```
